```python
#полностью весь код готов для расстановки локализаций

f = open("./dataset_pose/poses/00.txt", "r")

x_array = []
z_array = []

xz_array = []

for index, x in enumerate(f):
    x = x.split()
    value_x = x[3].replace(',', '.')
    x_array.append(float(value_x))
    value_z = x[11].replace(',', '.')
    z_array.append(float(value_z))
    
    xz_array.append((float(value_x), float(value_z), index))
    print(xz_array[index])
    
xz_array = np.array(xz_array)

xz_array = xz_array[xz_array[:, 0].argsort()]

print(len(xz_array))
print('x min = %f, x_max = %f, z_min = %f, z_max = %f' %(np.min(x_array), np.max(x_array), np.min(z_array), np.max(z_array)))

```

```python
lst_lst_size = 0
if np.min(x_array) < 0 and np.max(x_array)>=0 and (int(np.min(x_array))!=int(np.max(x_array))):
    lst_lst_size = abs(int(np.min(x_array)))+1 + int(np.max(x_array))+1
    
elif np.min(x_array) < 0 and np.max(x_array)<0 and (int(np.min(x_array))!=int(np.max(x_array))):
    lst_lst_size = abs(int(np.min(x_array))) - int(np.max(x_array))  +1
    
elif np.min(x_array) >= 0 and np.max(x_array)>=0 and (int(np.min(x_array))!=int(np.max(x_array))):
    lst_lst_size = abs(int(np.max(x_array))) - int(np.min(x_array))  +1

elif int(np.min(x_array))==int(np.max(x_array)) and int(np.min(x_array))!=0 and int(np.max(x_array))!=0:
    lst_lst_size = 1
    
elif int(np.min(x_array))==int(np.max(x_array)) and int(np.min(x_array))==0 and int(np.max(x_array))==0:
    if (np.min(x_array)<0 and np.max(x_array)<0) or (np.min(x_array)>0 and np.max(x_array)>0):
        lst_lst_size = 1
    else:
        lst_lst_size = 2
    
print(lst_lst_size)
```

```python
lst_lst = [[] for _ in range(lst_lst_size)]

print(len(lst_lst[0]))

lst = [xz_array[0]]
#lst_lst[0] = lst
#print(lst_lst[0][0])

index = 0
for idx, i in enumerate(xz_array):
    if len(lst_lst[index])==0:
        lst_lst[index].append(i)
        
    elif int(i[0]) == 0:
        if i[0] > (-1) and i[0]<0:
            print(index, i)
            lst_lst[index].append(i)
            if xz_array[idx+1][0] < 1 and xz_array[idx+1][0]>0:
                index+=1
                
        elif i[0] < 1 and i[0]>0:
            print(index, i)
            lst_lst[index].append(i)
                
            
    elif int(lst_lst[index][0][0]) == int(i[0]):
        lst_lst[index].append(i)
        if (idx+1)!=len(xz_array):
            if int(xz_array[idx+1][0]) == 0:
                index+=1
        
    else:
        index+=1
        lst_lst[index].append(i)

        
        
        
xz_array_new = np.zeros((len(xz_array), 5))

class_number = -1

count = 0

for i in lst_lst:
    if len(i)!=0:
        arr_sample = np.array(i)
        Y = arr_sample[arr_sample[:, 1].argsort()]
        print(Y)
        
        for idx, item in enumerate(Y):
            if idx == 0:
                class_number+=1
                sample = np.append(item, [class_number,0])
                xz_array_new[count] = sample
                
                count+=1

                
            #elif int(xz_array_new[count-1][1]) == 0:
            elif xz_array_new[count-1][1] > (-1) and xz_array_new[count-1][1]<0:
                sample = np.append(item, [class_number,0])
                xz_array_new[count] = sample
                count+=1

                if xz_array_new[count][1] > 0 and xz_array_new[count][1]<1:
                    class_number+=1

            elif int(xz_array_new[count-1][1]) < 1 and int(xz_array_new[count-1][1])>0:
                sample = np.append(item, [class_number,0])
                xz_array_new[count] = sample
                count+=1
                    
            elif int(xz_array_new[count-1][1]) == int(item[1]):
                sample = np.append(item, [class_number,0])
                xz_array_new[count] = sample
                count+=1
            else:
                class_number+=1
                sample = np.append(item, [class_number,0])
                xz_array_new[count] = sample
                count+=1
                
                
```
